```python
import pandas as pd
import numpy as np
```

```python
# loading data
path = "auto.csv"
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", "drive-wheels", "engine-location", "wheel-base", "length", "width", "height", "curb-weight", "engine-type", "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]

# Q1
df = pd.read_csv(path, names=headers)

# Q2
df.head(10)
```

```python
# Q3
df.replace("?", np.nan, inplace=True)

missing_data = df.isnull()
missing_data.head()

df.head(10)
```

```python
# Q4
missing_counts = {}

for column in missing_data.columns:
  missing_counts[column] = missing_data[column].value_counts().get(True, 0)

# missing_counts
```

```python
# Q5
# replace NaN by

replace_with_mean_cols = ["normalized-losses", "horsepower", "peak-rpm"]

for column in replace_with_mean_cols:
    # - I found that some data are not numbers (-> string)
    #   the mean method works only for numbers
    # - from string to number
    df[column] = pd.to_numeric(df[column])

    mean = df[column].mean()
    df[column].fillna(mean, inplace=True)

df.head()
```

```python
# Q6
# replace NaN with the most repeated value
df["num-of-doors"].fillna(df["num-of-doors"].value_counts().idxmax(), inplace=True)

df["num-of-doors"]
```

```python
# Q - Supprimez toute la ligne
df["price"].dropna(axis=0, inplace=True)

df["price"]
```

```python
# Q8 - Standardisation de données
# transform mpg to L/100km
# note: L/100km = 235 / mpg

df["city-mpg"] = 235/df["city-mpg"]
df.rename(columns={"city-mpg":"city-L/100km"}, inplace=True)
```

```python
df.describe()
```

```python
# Q9 - Normalisez les colonnes "length", "height" et "width".

# *by simple feature scaling
df['length'] = df['length']/df['length'].max()

# *by min-max
# df['length'] = (df['length']-df['length'].min())/(df['length'].max()-df['length'].min())

# *y z-score
# df['length'] = (df['length'] - df['length'].mean())/df['length'].std()

df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

```python
# Q10 - Regroupement de données
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)

group_name = ["Low", "Medium", "High"]

df['horsepower'] = pd.cut(df['horsepower'], bins, labels=group_name, include_lowest=True)
```

```python
# Q11 - Convertir des valeurs catégorielles en variables numériques (ou dummy variable)
new_aspiration = pd.get_dummies(df["aspiration"])

df.drop("aspiration", axis=1, inplace=True)

df = pd.concat([df, new_aspiration], axis=1)

df.head()
```

```python
# Q12 - export as csv
df.to_csv('output.csv', index=False)
```
